```python
class Rental():
    def __init__(self):
        self.rent = []
        self.down_payment = []
        self.expenses = []
        # self.closing = []
        self.rehab = []
        self.property = []
        self.cash_flow = []
        self.yearly_cash_flow = []
        self.total_investment = []
        self.r_o_i = []
        self.r_o_i1 = []


    def info(self):
        while True:
            property_cost = int(input('How much did you purchase the property for?'))
            self.property.append(property_cost)
            print(f'You have entered that you bought this property for ${property_cost}')
            rent_income = int(input('How much are you planning on charging for rent?'))
            # self.rent.append(rent_income)
            other_income = input('Are there any other forms of income on this property? (yes or no)').lower()
            if other_income == 'yes':
                other_incomes = int(input('How much do you make off of other forms of income on this property?'))
                self.rent.append(other_incomes + rent_income)
                # print(self.rent)
                break
            elif other_income == 'no':
                filler = 0
                self.rent.append(rent_income + filler)
                # # print(self.rent)
                break
            else:
                print('Please enter yes or no')
            # return other_income, other_incomes, property_cost

    def expense(self):
        while True:
            expense_costs = int(input("""What are your monthly expenses on this property? (Tax, Insurance, HOA, Morgage, etc...)
            Don't forget to include utilities if you are not expexting the tennant to cover those costs.
            It is recommended to include extra for potential repairs or property management fees.
            """))
            self.expenses.append(expense_costs)
            print(f'Your monthly expenses are ${expense_costs}')
            break

    def cash_flows(self):
        # raw_income = self.rent
        # raw_expenses = self.expenses
        for j in range(1):
            self.cash_flow.append(self.rent[j] - self.expenses[j])
            self.yearly_cash_flow = (self.cash_flow[j] * 12)
            self.cash_flow = str(self.cash_flow)[1:-1]
            
            print(f'Your current monthly cash flow is ${self.cash_flow}')
            print(f'Your yearly cash flow is ${self.yearly_cash_flow}')
            break

    def roi(self):
        down_p = int(input('What was your down payment on your property?'))
        closing_c = int(input('How much were the closing costs?'))
        rehab_c = input('Were there any rehab costs? (yes or no)').lower()
        for y in rehab_c:
            if rehab_c == 'yes':
                rehab_c1 = int(input('How much did you spend on rehab?'))
                self.rehab.append(rehab_c1)
                break
            else:
                filler = 0
                self.rehab.append(filler)
                pass
        misc_c = input('Were there any other misc costs? (yes or no)').lower()
        for y in misc_c:
            if misc_c == 'yes':
                misc_c1 = int(input('How much did you spend on misc?'))
                self.rehab.append(misc_c1)
                break
            else:
                pass
        self.total_investment.append(down_p + closing_c + rehab_c1)
        # self.total_investment = str(self.total_investment)[1:-1]
        # print(f'Your total investment is ${self.total_investment}')
        for x in range(1):
            self.r_o_i.append((self.yearly_cash_flow / self.total_investment[x]))
            self.r_o_i1.append(self.r_o_i[x] * 100 )
            self.r_o_i1 = str(self.r_o_i1)[1:-1]
            self.total_investment = str(self.total_investment)[1:-1]
            print(f'Your total investment is ${self.total_investment}')

            print(f'Your cash on cash ROI is {self.r_o_i1}%.')

        


            


                

property1 = Rental()
def runner():
    while True:
        property1.info()
        property1.expense()
        property1.cash_flows()
        property1.roi()
        break
    
runner()
```
